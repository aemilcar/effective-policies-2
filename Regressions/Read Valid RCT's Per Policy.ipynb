```python
import pandas as pd
import csv
```

```python
def read_data():
    rt = []
    with open('../../Data/valid_rcts_2014.csv') as f:
        reader = csv.reader(f)
        header = next(reader)
        for row in reader:
            l = [row[0], row[1]]
            s = row[2]
            s = s.split('[')[1]
            s = s.split(']')[0]
            s = s.split(', ')
            ids = [int(i) for i in s]
            l.append(ids)
            rt.append(l)
    return rt
```

```python
data = read_data()
```

```python
data
```

### Now, I need to create a dictionary
That maps the policy ID to valid RCT's

```python
policy_mapper = {}

for row in data:
    s1 = row[0]
    s2 = row[1]
    for p in row[2]:
        if p not in policy_mapper:
            policy_mapper[p] = {}
        policy_mapper[p][s1]=s2
```

```python
policy_mapper
```

```python
with open('../../Data/Policy_ID_RCT_Table.csv', 'w') as f:
    writer = csv.writer(f)
    writer.writerow(['PolicyID', 'Treatment State', 'Control State'])
    for policy in policy_mapper.keys():
        for s1, s2 in policy_mapper[policy].items():
            row = [policy, s1, s2]
            writer.writerow(row)
    
```
