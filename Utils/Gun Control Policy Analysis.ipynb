```python
df = df[ df["State"].isin(["Illinois", "New York"])]
```

```python
df.shape
```

```python
df.reset_index(inplace = True)
```

Let's print out the sum of each column and see where there's the biggest difference

```python
df.shape

iterations = list(range(0, df.shape[0], 2))
```

```python
import math

biggest_diff = 0
illinois_staging = []
new_york_staging = []

comparison_staging = {"Year":1000, "Column":"The most interesting one", "Difference" : 0}

for i in iterations:
    
    ILLINOIS = i
    NEW_YORK = i+1
        
    # walk through columns and get the comparator
    for idx in range(len(list(df))):
        il = df.iloc[ILLINOIS][idx]
        ny = df.iloc[NEW_YORK][idx]
                
        if (idx < 3) or (math.isnan(il)) or (math.isnan(ny)):
            continue
        
        diff = abs(il - ny)
        if diff > biggest_diff:
            print ("Triggered change")
            comparison_staging["Year"] = df.iloc[ILLINOIS]["Year"]
            comparison_staging["Column"] = list(df)[idx]
            comparison_staging["Difference"] = diff
            biggest_diff = diff
            print ("Biggest diff reset to {}".format(biggest_diff))
                

        
        
```
